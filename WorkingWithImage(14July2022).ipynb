# Working with image 

```python
from PIL import Image 
```

```python
mac = Image.open('example..jpg')
```

# Cropping Image

```python
mac.crop((0,0,100,100))
```

# pasting image 

```python
patter = mac.crop((10,10,250,250))
```

```python
mac.paste(im=patter,box=(100,0))
```

```python
mac
```

# resize

```python
mac.resize((300,500))
```

```python
mac.rotate(90)
```

# Color Transpiracy

```python
red = Image.open('red.jpg')
```

```python
red.putalpha(100)
red
```

```python
blue = Image.open('blue.png')
```

```python
blue.putalpha(100)
blue
```

```python
green = Image.open('green.jpg')
```

```python
green.putalpha(100)
green
```

# Mixing the colors 

```python
blue.paste(im=red,box=(0,0),mask = red)
```
